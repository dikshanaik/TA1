```
import numpy as np
import re
import scipy as sp
import pandas as pd
import sklearn as sk
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

```
data = pd.read_csv("https://raw.githubusercontent.com/AnoVando/MSIS/master/IA3.csv")

data.head()
```

```
data.isnull().any()
```

```
reviewData = data["review"].tolist()
```

```
#Tokanize Step 1
token_review =[]
for i in reviewData:
  #print(i)
  token= nltk.word_tokenize(i)
  token_review.append(token)

print(token_review)
```

```
print(token_review[0])
```

```
#Lemmetize Step 1
lem_review = []
lemmatizer = nltk.stem.WordNetLemmatizer()
for i in token_review:
  lemmatized_review = [lemmatizer.lemmatize(x) for x in i if x.isalpha()]
  lem_review.append(lemmatized_review)
print(lem_review)
```

```
print(lem_review[0])
```

```
#remove stopwords Step 1
from nltk.corpus import stopwords
stopWordEnglish = stopwords.words('english')
stopword_list=['I','l','lt','gt','quot','wa','ha','Ourselves','Hers','Between','Yourself','But','Again','There','About','Once',
               'During','Out','Very','Having','With','They','Own','an','Be','Some','For','Do', 'Its', 'Yours', 'Such', 'Into',
               'Of', 'Most', 'Itself', 'Other', 'Off', 'Is', 'S', 'Am', 'Or', 'Who', 'As', 'From', 'Him', 'Each', 'The', 'Themselves',
               'Until', 'Below', 'Are', 'We', 'These', 'Your', 'His', 'Through', 'Don', 'Nor', 'Me', 'Were', 'Her', 'More', 'Himself',
               'This', 'Down', 'Should', 'Our', 'Their', 'While', 'Above', 'Both', 'Up', 'To', 'Ours', 'Had', 'She', 'All', 'No', 'When',
               'At', 'Any', 'Before', 'Them', 'Same', 'And', 'Been', 'Have', 'In', 'Will', 'On', 'Does', 'Yourselves', 'Then', 'That', 'Because',
               'What', 'Over', 'Why', 'So', 'Can', 'Did', 'Not', 'Now', 'Under', 'He', 'You', 'Herself', 'Has', 'Just', 'Where', 'Too', 'Only',
               'Myself', 'which', 'those', 'After', 'Few', 'Whom', 'T', 'Being', 'If', 'Theirs', 'My', 'Against', 'A', 'By', 'Doing', 'It', 'How',
               'Further', 'Was', 'Here', 'Than', 'Also','also']
stopWordEnglish.extend(stopword_list)
stop_words = []

for stop in lem_review:
  stop_words_removed = [x for x in stop if x not in stopWordEnglish]
  stop_words.append(stop_words_removed)
  
print(stop_words)
```

```
###Vectorize with Bigram Step 1 
from sklearn.feature_extraction.text import CountVectorizer

input2 = [" ".join(x) for x in stop_words]

count = CountVectorizer(ngram_range=(2, 2), min_df=5)
count.fit(input2)
v2 = count.transform(input2)
terms = count.get_feature_names()
print(terms)
print(count.vocabulary_)
print('\n')
print(v2.toarray())
```

```
#Step 2 LDA with 6 topics
from sklearn.decomposition import LatentDirichletAllocation
lda = LatentDirichletAllocation(n_components=6).fit(v2)

for topic_idx,topic in enumerate(lda.components_):
    print(f'Topic #{topic_idx}:')
    print(" ".join([terms[i] for i in topic.argsort()[:-6-1:-1]]))
    print('\n')
```

```
#Step 3 Food review topic distribution with top-2 topics
f1=lda.transform(v2[1])
f2=lda.transform(v2[2])
f3=lda.transform(v2[3])
f4=lda.transform(v2[4])
f5=lda.transform(v2[5])
f6=lda.transform(v2[6])
f7=lda.transform(v2[7])
f8=lda.transform(v2[8])
f9=lda.transform(v2[9])
f10=lda.transform(v2[10])
print(f1)
print(f2)
print(f3)
print(f4)
print(f5)
print(f6)
print(f7)
print(f8)
print(f9)
print(f10)
```

**Based on the topic distributions, the top 2 topics are 3 & 6**

```
#Step 4 using Unigram to get 5 words
input3 = [" ".join(x) for x in stop_words]

#print(input3)

top5 = CountVectorizer(ngram_range=(1, 1), min_df=5)
top5.fit(input3)
v3 = top5.transform(input3)
features = top5.get_feature_names()
```

```
lda_top5 = LatentDirichletAllocation(n_components=6).fit(v3)

for topic_idx,topic in enumerate(lda_top5.components_):
    print(f'Topic #{topic_idx}:')
    print(" ".join([features[i] for i in topic.argsort()[:-5-1:-1]]))
    print('\n')
```

**Topic 1 is mostly a movie review about a love story film, 
Topic 2 is mostly a food review about cake & coffee, 
Topic 3 is mostly a food review about delicious food, 
Topic 4 is mostly a movie review about love film, 
Topic 5 is mostly a movie review about woman oriented or book inspired film, 
Topic 6 is mostly a movie review about war film**
